```python
import pandas as pd
import numpy as np
np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
```

```python
sales = pd.read_csv("sales.csv")
sales.head()
```

```python
sales['Revenue'].plot(kind = 'kde')
```

```python
sales.groupby('Order method type').sum().plot(kind = 'bar', y = 'Revenue')
```

```python
sales.groupby('Year').sum().plot(kind = 'bar', y = 'Revenue')
```

```python
sales.groupby('Year')[['Revenue', 'Planned revenue','Gross profit']].plot(kind= 'box')
```
